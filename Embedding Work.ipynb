## Reading Embedding Files

```python
from gensim.models import KeyedVectors
w2v = KeyedVectors.load_word2vec_format("/home/gauravroy/Desktop/intent/GoogleNews-vectors-negative300.bin",binary = True, limit = 500000)
```

```python
import numpy as np
def loadGloveModel(gloveFile):
    print ("Loading Glove Model")
    f = open(gloveFile,'r')
    model = {}
    for line in f:
        splitLine = line.split()
        word = splitLine[0]
    try:
        embedding = np.array([float(val) for val in splitLine[1:]])
        model[word] = embedding
        print ("Done.") 
        print(len(model)) 
        print(" words loaded!")
        return (model)
    except ValueError :
            print("error")

glove = loadGloveModel("/home/gauravroy/Desktop/intent/glove.840B.300d.txt")
```

## Reading Dictionaries

```python
import pickle

with open('/home/gauravroy/Desktop/slotfilling/slot_word_dict.pickle', 'rb') as fp:
    dicts = pickle.load(fp)
fp.close()

slot_dict,word_dict = dicts
```

## Storing Embeddings for required words

```python
glove_embeddings = dict()
for word in word_dict:
    try:
        glove_embeddings[word] = glove[word]
    except:
        ;

w2v_embeddings = dict()
for word in word_dict:
    try:
        w2v_embeddings[word] = w2v[word]
    except:
        ;
```

```python
with open('/home/gauravroy/Desktop/slotfilling/glove_embeddings.pickle', 'wb') as fp:
     pickle.dump(glove_embeddings, fp)
fp.close()

with open('/home/gauravroy/Desktop/slotfilling/w2v_embeddings.pickle', 'wb') as fp:
     pickle.dump(w2v_embeddings, fp)
fp.close()
```

## Embedding Matrix Combined

```python
nb_words = len(word_dict)
EMBEDDING_DIM = 600
embedding_matrix = np.zeros((nb_words+1, EMBEDDING_DIM))

i = 0
x_g = x_w = 0
for word in word_dict:
    try:
        embedding_matrix[word_dict[word]][:300] = glove_embeddings[word]
    except KeyError:
        x_g += 1
    try:
        embedding_matrix[word_dict[word]][300:] = w2v_embeddings[word]
    except KeyError:
        x_w += 1
```

```python
with open('/home/gauravroy/Desktop/slotfilling/embedding_matrix.pickle', 'wb') as fp:
     pickle.dump(embedding_matrix,fp)
fp.close()
```
